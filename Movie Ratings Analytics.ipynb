```python
##for plots present in seaborn 
sns.set_style("darkgrid")
f,axes=plt.subplots(2,2,figsize=(12,6))
k1=sns.kdeplot(movies.Budget,movies.AudienceRatings,ax=axes[0,0])
k2=sns.kdeplot(movies.Budget,movies.CriticsRatings,ax=axes[0,1])
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
sns.violinplot(data=movies[movies.Genre=='Drama'],y='CriticsRatings',x='Year',ax=axes[1,0])
##sns.regplot(data=movies,x='CriticsRatings',y='AudienceRatings',fit_reg=False,ax=axes[1,1]) instead of lmplot\
## for dashboard use regplot
##axes[1,1].hist(movies[movies.Genre =='Action'].Budget,bins=17) for plots not present in seaborn
k4=sns.kdeplot(movies.CriticsRatings,movies.AudienceRatings,shade=True,shade_lowest=False,cmap='Reds',ax=axes[1,1])
k4b=sns.kdeplot(movies.CriticsRatings,movies.AudienceRatings,cmap='Reds',ax=axes[1,1]) #two graphs can be overlapped 
plt.show()
```

```python
##styling up the dashboard
sns.set_style("dark",{"axes.facecolor":"black"}) #called a dictionary for black background
f,axes=plt.subplots(2,2,figsize=(18,9))
#plot[0,0]
k1=sns.kdeplot(movies.Budget,movies.AudienceRatings,ax=axes[0,0],shade=True,shade_lowest=True,cmap='inferno')
k1b=sns.kdeplot(movies.Budget,movies.AudienceRatings,ax=axes[0,0],cmap='cool')
#plot[0,1]
k2=sns.kdeplot(movies.Budget,movies.CriticsRatings,ax=axes[0,1],shade=True,shade_lowest=True,cmap='inferno')
k2=sns.kdeplot(movies.Budget,movies.CriticsRatings,ax=axes[0,1],cmap='cool')
#coordinates
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
#plot[1,0]
sns.violinplot(data=movies,y='Budget',x='Year',ax=axes[1,0],palette='YlOrRd')
#plot[1,1]
k4=sns.kdeplot(movies.CriticsRatings,movies.AudienceRatings,shade=True,shade_lowest=False,cmap='Blues',ax=axes[1,1])
k4b=sns.kdeplot(movies.CriticsRatings,movies.AudienceRatings,cmap='gist_gray_r',ax=axes[1,1]) #two graphs can be overlapped 
plt.show()
```

```python
list1=list()
mylabels=[]
for genr in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre ==genr].Budget)
    mylabels.append(genr)
sns.set_style("whitegrid")
fig,ax=plt.subplots()
fig.set_size_inches(11.7,8.27)
plt.title("Movie Budget Distribution",fontsize=30,color='Darkblue',fontname='Arial')
plt.ylabel('Number of Movies',fontsize=25,color='Red')
plt.xlabel('Budget',fontsize=25,color='Green')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.hist(list1,bins=30,stacked=True, rwidth=1,label=mylabels)
plt.legend(frameon=True ,fancybox=True ,framealpha=1,shadow=True,prop={'size':20})
plt.show()
```
